Um die Filme aus der Datenbank abzurufen, benötige ich eine Reihe von zufälligen IDs, die ich in einer Liste speichern werde. Diese IDs werden mithilfe der Zufallsmethode von Numpy ermittelt.

Ich habe beschlossen, mindestens 10000 Identifikatoren zu generieren, falls sich einige Identifikatoren in der Liste als ungültig erweisen sollten. Außerdem möchte ich vermeiden, dass einige der durchsuchten Filme nicht die Informationen enthalten, die ich für meine Analyse benötige, insbesondere Informationen über das Budget und die Einnahmen.

```python
np.random.seed(0)
movieId = np.random.randint(10, 3000, [10000,])
```

# Try and Except

Um zu verhindern, dass die Methode im Fehlerfall gestoppt wird, d. h. wenn eine ID in der Liste nicht existiert, habe ich try and except verwendet.

```python
imdbValidId = []
for movieId in listeMovieid:
    movie = tmdb.Movies(movieId)
    try:
        response = movie.info()
        id = response['id']
        imdbValidId.append(id)
        print(id)
    except:
        print('NaN')
```

# Gültige IDs

# Entfernung von wiederholten IDs

es kann sein die Methode mehrmals gleiche numbers erstellt hat

Um zu verhindern, dass ein und derselbe Film mehrmals genommen wird, gehe ich die Liste durch, um sich wiederholende IDs zu löschen.

```python
imdbValidId_final = []

for ids in imdbValidId:
    
    if ids in imdbValidId_final:
        print('schon')
        
    else: 
        imdbValidId_final.append(ids)
```

```python
len(imdbValidId_final)
```

# Abruf von attributen und Erstellung von Datensatz

Ich werde die Elemente der Liste durchsuchen und die oben im Code definierten Attribute abrufen.

Ich werde durch gültige IDs, die ich erhalten habe, Informationen abrufen, die ich brauche

Um dies zu erreichen, habe ich eine for-Schleife erstellt und die wiederhergestellten Elemente in einer Liste gespeichert und anschließend die Elemente aus der Liste wiederhergestellt und einen Datensatz angelegt.

```python
movies = []
id = []
title = []
genres = []
time = []
revenue = []
release_date = []
popularity = []
budget = []
actors =[]

for movies in imdbValidId_final:
    movie = tmdb.Movies(movies)
    response1 = movie.info()
    response2 = movie.credits()
    
    movieid = response1['id']    
    id.append(movieid)
    
    movietitle = response1['title']
    title.append(movietitle)
    
    moviebudget = response1['budget']
    budget.append(moviebudget)
    
    movietime = response1['runtime']
    time.append(movietime)
    
    movierevenue = response1['revenue']
    revenue.append(movierevenue)
    
    moviepopularity = response1['popularity']
    popularity.append(moviepopularity)
    
    moviedate = response1['release_date']
    release_date.append(moviedate)
    
    moviegenres = response1['genres']
    
    genrename = []
    for genre in moviegenres:
        moviegenre = genre['name']
        genrename.append(moviegenre)
        
    str_genres= ', '.join(genrename)
    genres.append(str_genres)
    
    actorsname = response2['cast']
    
    celebrities =[]
    for name in actorsname:
        celebrity = name['name']
        celebrities.append(celebrity)
        
    str_actors = ', '.join(celebrities)
    actors.append(str_actors)
    
    
    df = pd.DataFrame({
        
        'id': id,
        'Titel': title,
        'Genres': genres,
        'Erscheinungsdatum': release_date,
        'Spieldauer': time,
        'Budget': budget,
        'Einnahmen': revenue,
        'Popularitaet': popularity,
        'Schauspieler': actors
    })
    
    print('sucess')

df.to_csv('imdbDatensatz.csv', encoding='utf-8', index=False)

    
```

# Frage 3: Ist die Popularität  eines Films von seiner Dauer abhängig?
